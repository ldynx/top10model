```python
jam = pandas.read_csv("/home/songyeji/다운로드/jam.csv")
jam.columns
```

```python
prc = pandas.read_csv("/home/songyeji/다운로드/prc_light_ver.csv", low_memory = False)
prc.columns
```

```python
# i년도 j분기(1,4,7,10월)에 대한 데이터

i=2008
j=1
date = i*100+j

data_10 = pandas.read_csv("/home/songyeji/다운로드/시가총액순위/"+str(date)+".csv")
data_10.columns
```

```python
from pandas import Series, DataFrame

data=pandas.DataFrame()
stock=pandas.DataFrame()

if j==1:
    myym = date-100+11
else:
    myym = date-1

for name in data_10['종목코드']:
    row = DataFrame(jam[jam['code'].isin([name])])
    row = row[row['ym'].isin([myym])]
    row = row[row['gb'].isin(['1','2','3','4'])]
    row2 = DataFrame(prc[prc['stk_cd'].isin([str(name)])])
    row2 = row2[row2['ymd'].isin(range((myym-2)*100, (myym-2)*100+300))]
    data = data.append(row, ignore_index=True)
    stock = stock.append(row2, ignore_index=True)
```

```python
stock
```

```python
# 전체 발행 주식 수totStocks, 시가총액capital 월별 평균

totStocks = Series([])
capital = Series([])

for name in data_10['종목코드']:
    row3 = stock[stock['stk_cd'].isin([str(name)])]
    amount_stock = row3['stks'].sum() / len(row3)
    amount_price = row3['cprc1'].sum() / len(row3)
    amount_capital = amount_stock * amount_price
    totStocks = totStocks.append(Series([amount_stock]), ignore_index=True)
    capital = capital.append(Series([amount_capital]), ignore_index=True)
    
capital = capital/1000
name = data_10['종목코드']

prc_result = pandas.DataFrame({
    'name' : name,
    'totstocks' : totStocks,
    'capital' : capital
})
```

```python
# 분기별 인덱스
# 유동비율cRatio, 부채비율dRatio, 이자보상배율IntrstRatio, 영업이익률OperPercentage, ROA, ROE, BPS, EPS, PBS

result = pandas.DataFrame({
    'name' : name,
   'cRatio' : data['cAssets']/ data['cLiabilities'],
    'dRatio' : data['liabilities']/ data['equity'],
    'IntrstRatio' : data['opProfit']/ data['IntrstExpense'],
    'OperPercentage' : data['opProfit']/ data['netSales'],
    'ROA' : data['netProfit']/ data['assets'],
    'ROE' : data['netProfit']/ data['equity'],
    'BPS' : data['equity']/ totStocks*1000,
    'EPS' : data['netProfit']/ totStocks*1000,
    'PER' : capital/ data['netProfit'],
    'PBR' : capital/ data['equity'],
})

result
```
